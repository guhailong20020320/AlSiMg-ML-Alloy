```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm


# 设置中文字体和数学符号显示
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['axes.unicode_minus'] = False

import warnings
# 忽略所有警告
warnings.filterwarnings("ignore")

# 读取数据
path = r"AlSiMgTiSrSc成分.xlsx"
df = pd.read_excel(path)

# 选择数值列计算相关性
df_numeric = df.select_dtypes(include=np.number)
corr_matrix = df_numeric.corr()

columns = corr_matrix.columns
n_cols = len(columns)

# 创建画布
fig, ax = plt.subplots(figsize=(16, 13))

# 设置颜色映射（从红到蓝，适合表达正负相关性）
cmap = plt.cm.get_cmap('Spectral_r', 20)
norm = plt.Normalize(vmin=-1, vmax=1)  # 映射范围 -1 到 1

# 圆圈大小系数、字体等视觉参数
circle_multiplier = 3000
text_fontsize = 18  
shadow_offset_x = 0.05  # 阴影 x 偏移
shadow_offset_y = -0.05  # 阴影 y 偏移

# 遍历相关系数矩阵
for i in range(n_cols):
    for j in range(n_cols):
        corr_val = corr_matrix.iloc[i, j]  # 相关系数值
        abs_corr_val = abs(corr_val)
        
        # 不显示对角线
        if i == j:
            continue
        
        # 左下角区域 i > j：绘制圆圈表示相关性
        if i > j:
            color = cmap(norm(corr_val))  # 映射颜色
            size = abs_corr_val * circle_multiplier  # 映射大小
            ax.scatter(j, i, s=size, color=color, alpha=0.8, 
                      edgecolors='black', linewidth=0.5, zorder=3)
        
        # 右上角区域 i < j：绘制带阴影的长方形 + 显示数值
        elif i < j:
            # 计算矩形宽度，反映相关性强弱
            min_rect_width = 0.7    # 最小宽度
            max_rect_width_ratio = 1    # 最大宽度
            rect_width = min_rect_width + abs_corr_val * (max_rect_width_ratio - min_rect_width)
            rect_width = min(rect_width, max_rect_width_ratio)
            
            # 使矩形居中
            rect_x = j - rect_width / 2
            rect_y = i - 0.45
            
            # 阴影矩形
            shadow_rect = plt.Rectangle(
                (rect_x + shadow_offset_x, rect_y + shadow_offset_y),
                rect_width, 0.9,
                facecolor='gray', edgecolor='none', alpha=0.4, zorder=0
            )
            ax.add_patch(shadow_rect)
            
            # 前景矩形
            main_rect = plt.Rectangle(
                (rect_x, rect_y), rect_width, 0.9,
                facecolor='white', edgecolor='black', linewidth=0.5,
                alpha=0.8, zorder=1
            )
            ax.add_patch(main_rect)
            
            # 文本颜色根据相关性映射
            text_color = cmap(norm(corr_val))
            
            # 显示相关系数文字
            ax.text(j, i, f'{corr_val:.2f}',
                    ha='center', va='center',
                    fontsize=text_fontsize, fontweight='bold',
                    color=text_color, zorder=2)

# 设置坐标轴标签、刻度字体
ax.set_xticks(np.arange(n_cols))
ax.set_yticks(np.arange(n_cols))

ax.set_xticklabels(columns, rotation=45, ha='right',
                   fontsize=18, fontweight='bold')
ax.set_yticklabels(columns,
                   fontsize=18, fontweight='bold')

# 坐标范围
ax.set_xlim(-0.5, n_cols - 0.5)
ax.set_ylim(n_cols - 0.5, -0.5)

ax.grid(False)
ax.set_aspect('equal', adjustable='box')

# 添加颜色条（Colorbar）
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, shrink=0.7)

# 设置坐标轴刻度和样式
font_prop = fm.FontProperties(family='Times New Roman', weight='bold', size=40)
plt.xticks(fontproperties=font_prop)
plt.yticks(fontproperties=font_prop)

plt.tick_params(axis='both', direction='in')
ax = plt.gca()
ax.spines['bottom'].set_linewidth(5)
ax.spines['left'].set_linewidth(5)
ax.spines['top'].set_linewidth(5)
ax.spines['right'].set_linewidth(5)
ax.tick_params(axis='both', which='major', width=4)
ax.tick_params(axis='both', which='minor', width=2)

# 设置刻度线长度（关键参数）
ax.tick_params(axis='both', which='major', length=8)  # 主刻度线加长到10


# 调整颜色条刻度字体大小和加粗
cbar.ax.tick_params(labelsize=20)
for tick in cbar.ax.get_yticklabels():
    tick.set_fontweight('bold')

# 保存和显示
plt.tight_layout()
plt.savefig('皮尔逊相关图_上三角——Spectral_r-1200.tif', format='tif', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import shap
import matplotlib.font_manager as fm  # 修正导入
from sklearn.model_selection import train_test_split
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# ======= 1. 加载数据 =======
dataset = pd.read_excel('AlSiMgTiSrSc.6.26.4.xlsx')

# 特征和标签列
features = ['Si', 'Mg', 'Zn', 'Cu', 'Mn', 'Fe', 'Ti', 'Sr', 'Sc']
label_column = 'UTS'

# 特征和标签分离
X = dataset[features]
y = dataset[label_column]

# ======= 2. 拆分数据集 =======
x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=55
)

# ======= 3. 训练 CatBoost 模型 =======
model = CatBoostRegressor(
    depth=6, 
    iterations=488, 
    l2_leaf_reg=2, 
    learning_rate=0.1, 
    random_seed=67,
    verbose=0
)
model.fit(x_train, y_train)

# ======= 4. 计算 SHAP 值 =======
# 创建 SHAP 解释器
explainer = shap.Explainer(model, x_train)
shap_values = explainer(x_train)

# 计算平均 SHAP 值（绝对值）
mean_shap_value1 = np.abs(shap_values.values).mean(0)

# ======= 5. 数据准备 =======
# 创建 SHAP 汇总 DataFrame
shap_summary = pd.DataFrame({
    "Feature": features,
    "MeanSHAP": mean_shap_value1
}).sort_values("MeanSHAP", ascending=False)

# 获取排序后的特征
sorted_feats = shap_summary["Feature"].values

# 计算 SHAP 平均值与百分比
mean_shap = shap_summary["MeanSHAP"].values
mean_shap_pct = mean_shap / mean_shap.sum() * 100

# ======= 6. 绘制 beeswarm 图 =======
# 设置字体
font_prop = fm.FontProperties(family='Times New Roman', weight='bold', size=18)

# 创建总结图    #配色   viridis  Spectral   coolwarm  RdYlGn  RdYlBu  RdBu  RdGy  PuOr  BrBG PRGn  PiYG 
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values.values, x_train, show=False, 
                  feature_names=features, plot_type='dot', cmap='coolwarm')

# 获取当前轴对象
ax = plt.gca()

# 调整刻度线朝内
ax.tick_params(axis='both', direction='in', length=4)

# 设置坐标轴刻度和样式
plt.xticks(fontproperties=font_prop, fontsize=18)
plt.yticks(fontproperties=font_prop, fontsize=18)

# 设置坐标轴标题
ax.set_xlabel("SHAP value (Impact on model output)", fontproperties=font_prop, fontsize=22)
ax.set_ylabel(ax.get_ylabel(), fontproperties=font_prop, fontsize=22)

# 设置坐标轴线宽
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.tick_params(axis='both', which='major', width=1.5)
ax.tick_params(axis='both', which='minor', width=1.0)

# 获取colorbar对象
fig = plt.gcf()
all_axes = fig.get_axes()

# 修改colorbar的字体
if len(all_axes) > 1:
    cbar = all_axes[-1]  # colorbar通常是最后一个axes
    cbar.set_ylabel(cbar.get_ylabel(), fontproperties=font_prop, fontsize=18)
    cbar.tick_params(labelsize=18)
    for label in cbar.get_yticklabels():
        label.set_fontproperties(font_prop)
        
plt.tight_layout()
plt.savefig('SHAP_蜂窝图_UTS-1200.tif', dpi=1200, bbox_inches='tight')
plt.show()

# ======= 7. 绘制带百分比的柱状图 =======
plt.figure(figsize=(12, 8), dpi=1200)

# 设置配色方案
cmap_name = "coolwarm"
cmap1 = plt.get_cmap(cmap_name)
colors = cmap1(np.linspace(0, 1, len(shap_summary)))

# 绘制水平柱状图
plt.barh(range(len(shap_summary)), shap_summary["MeanSHAP"], 
         height=0.6, alpha=0.6, color=colors, edgecolor="none")

# 设置柱状图横坐标范围
bar_xlim = max(shap_summary["MeanSHAP"]) * 1.1 if max(shap_summary["MeanSHAP"]) > 0 else 1.0
plt.xlim(0, bar_xlim)
plt.xticks(np.linspace(0, bar_xlim, 7), 
           [f"{x:.2f}" for x in np.linspace(0, bar_xlim, 7)], fontsize=11)
plt.xlabel("Mean(|SHAP| value)", fontsize=22, fontweight='bold')

# 设置 y 轴
plt.yticks(range(len(shap_summary)), shap_summary["Feature"], fontsize=11)
plt.ylim(-0.5, len(shap_summary)-0.5)

# 在柱状图旁显示百分比
for i, (shap_val, pct) in enumerate(zip(shap_summary["MeanSHAP"], mean_shap_pct)):
    plt.text(shap_val * 0.12, i, f"({pct:.1f}%)", 
             va="center", ha="left", fontsize=10, color="black", weight='bold')

plt.title('SHAP Feature Importance with Percentage', fontsize=22, fontweight='bold')
plt.gca().invert_yaxis()  # 最重要的特征在顶部

# 调整刻度线朝内
ax = plt.gca()
ax.tick_params(axis='both', direction='in', length=4)

# 设置坐标轴刻度和样式
font_prop = fm.FontProperties(family='Times New Roman', weight='bold', size=18)
plt.xticks(fontproperties=font_prop)
plt.yticks(fontproperties=font_prop)

# 设置坐标轴线宽和刻度线宽度
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.tick_params(axis='both', which='major', width=1.5)
ax.tick_params(axis='both', which='minor', width=1.0)

# 保存柱状图
plt.tight_layout()
plt.savefig('SHAP_权重百分比图_UTS-1200.tif', dpi=1200, bbox_inches='tight')
plt.show()

# ======= 8. 计算模型性能指标 =======
# 预测
y_pred = model.predict(x_test)

# 计算指标
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\n" + "="*50)
print("模型性能指标:")
print("="*50)
print(f"均方误差 (MSE): {mse:.4f}")
print(f"均方根误差 (RMSE): {rmse:.4f}")
print(f"平均绝对误差 (MAE): {mae:.4f}")
print(f"决定系数 (R²): {r2:.4f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import shap
import matplotlib.font_manager as fm  # 修正导入
from sklearn.model_selection import train_test_split
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# ======= 1. 加载数据 =======
dataset = pd.read_excel('AlSiMgTiSrSc.6.26.4.xlsx')

# 特征和标签列
features = ['Si', 'Mg', 'Zn', 'Cu', 'Mn', 'Fe', 'Ti', 'Sr', 'Sc']
label_column = 'EL'

# 特征和标签分离
X = dataset[features]
y = dataset[label_column]

# ======= 2. 拆分数据集 =======
x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=9
)

# ======= 3. 训练 CatBoost 模型 =======
model = CatBoostRegressor(
    depth=9, 
    iterations=449, 
    l2_leaf_reg=3.078219166921273, 
    learning_rate=0.1, 
    random_seed=73,
    verbose=0
)
model.fit(x_train, y_train)

# ======= 4. 计算 SHAP 值 =======
# 创建 SHAP 解释器
explainer = shap.Explainer(model, x_train)
shap_values = explainer(x_train)

# 计算平均 SHAP 值（绝对值）
mean_shap_value1 = np.abs(shap_values.values).mean(0)

# ======= 5. 数据准备 =======
# 创建 SHAP 汇总 DataFrame
shap_summary = pd.DataFrame({
    "Feature": features,
    "MeanSHAP": mean_shap_value1
}).sort_values("MeanSHAP", ascending=False)

# 获取排序后的特征
sorted_feats = shap_summary["Feature"].values

# 计算 SHAP 平均值与百分比
mean_shap = shap_summary["MeanSHAP"].values
mean_shap_pct = mean_shap / mean_shap.sum() * 100

# ======= 6. 绘制 beeswarm 图 =======
# 设置字体
font_prop = fm.FontProperties(family='Times New Roman', weight='bold', size=18)

# 创建总结图    #配色   viridis  Spectral   coolwarm  RdYlGn  RdYlBu  RdBu  # RdGy  PuOr  BrBG PRGn  PiYG  #PRGn
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values.values, x_train, show=False, 
                  feature_names=features, plot_type='dot', cmap='PRGn')

# 获取当前轴对象
ax = plt.gca()

# 调整刻度线朝内
ax.tick_params(axis='both', direction='in', length=4)

# 设置坐标轴刻度和样式
plt.xticks(fontproperties=font_prop, fontsize=18)
plt.yticks(fontproperties=font_prop, fontsize=18)

# 设置坐标轴标题
ax.set_xlabel("SHAP value (Impact on model output)", fontproperties=font_prop, fontsize=22)
ax.set_ylabel(ax.get_ylabel(), fontproperties=font_prop, fontsize=22)

# 设置坐标轴线宽
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.tick_params(axis='both', which='major', width=1.5)
ax.tick_params(axis='both', which='minor', width=1.0)

# 获取colorbar对象
fig = plt.gcf()
all_axes = fig.get_axes()

# 修改colorbar的字体
if len(all_axes) > 1:
    cbar = all_axes[-1]  # colorbar通常是最后一个axes
    cbar.set_ylabel(cbar.get_ylabel(), fontproperties=font_prop, fontsize=18)
    cbar.tick_params(labelsize=18)
    for label in cbar.get_yticklabels():
        label.set_fontproperties(font_prop)
        
plt.tight_layout()
plt.savefig('SHAP_蜂窝图_EL-1200.tif', dpi=1200, bbox_inches='tight')
plt.show()

# ======= 7. 绘制带百分比的柱状图 =======
plt.figure(figsize=(12, 8), dpi=1200)

# 设置配色方案
cmap_name = "PRGn"
cmap1 = plt.get_cmap(cmap_name)
colors = cmap1(np.linspace(0, 1, len(shap_summary)))

# 绘制水平柱状图
plt.barh(range(len(shap_summary)), shap_summary["MeanSHAP"], 
         height=0.6, alpha=0.6, color=colors, edgecolor="none")

# 设置柱状图横坐标范围
bar_xlim = max(shap_summary["MeanSHAP"]) * 1.1 if max(shap_summary["MeanSHAP"]) > 0 else 1.0
plt.xlim(0, bar_xlim)
plt.xticks(np.linspace(0, bar_xlim, 7), 
           [f"{x:.2f}" for x in np.linspace(0, bar_xlim, 7)], fontsize=11)
plt.xlabel("Mean(|SHAP| value)", fontsize=22, fontweight='bold')

# 设置 y 轴
plt.yticks(range(len(shap_summary)), shap_summary["Feature"], fontsize=11)
plt.ylim(-0.5, len(shap_summary)-0.5)

# 在柱状图旁显示百分比
for i, (shap_val, pct) in enumerate(zip(shap_summary["MeanSHAP"], mean_shap_pct)):
    plt.text(shap_val * 0.12, i, f"({pct:.1f}%)", 
             va="center", ha="left", fontsize=10, color="black", weight='bold')

plt.title('SHAP Feature Importance with Percentage', fontsize=22, fontweight='bold')
plt.gca().invert_yaxis()  # 最重要的特征在顶部

# 调整刻度线朝内
ax = plt.gca()
ax.tick_params(axis='both', direction='in', length=4)

# 设置坐标轴刻度和样式
font_prop = fm.FontProperties(family='Times New Roman', weight='bold', size=18)
plt.xticks(fontproperties=font_prop)
plt.yticks(fontproperties=font_prop)

# 设置坐标轴线宽和刻度线宽度
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.tick_params(axis='both', which='major', width=1.5)
ax.tick_params(axis='both', which='minor', width=1.0)

# 保存柱状图
plt.tight_layout()
plt.savefig('SHAP_权重百分比图_EL-1200.tif', dpi=1200, bbox_inches='tight')
plt.show()

# ======= 8. 计算模型性能指标 =======
# 预测
y_pred = model.predict(x_test)

# 计算指标
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\n" + "="*50)
print("模型性能指标:")
print("="*50)
print(f"均方误差 (MSE): {mse:.4f}")
print(f"均方根误差 (RMSE): {rmse:.4f}")
print(f"平均绝对误差 (MAE): {mae:.4f}")
print(f"决定系数 (R²): {r2:.4f}")
```
